```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import collections
amazon_jobs = pd.read_csv('amazon_jobs_dataset.csv')
df = amazon_jobs.copy()
df
```

```python
posting_date = df.Posting_date.tolist()
posting_date
```

```python
dates = []
for i in posting_date:
    length = len(i)
    dates.append(int(i[length - 4: length]))
dates
```

```python
mapping = {}
for i in dates:
    if i in mapping:
        mapping[i] += 1
    else:
        mapping[i] = 1
mapping = collections.OrderedDict(sorted(mapping.items()))
mapping
```

```python
year = []
job_posting = []
for i in mapping:
    year.append(i)
    job_posting.append(mapping[i])
plt.plot(year, job_posting)
plt.show()
```

```python
for i in mapping:
    print(i, mapping[i])
```

```python
months = []
for i in posting_date:
    months.append(i.split(" ")[0])
months
```

```python
mapping = {}
for i in months:
    if i in mapping:
        mapping[i] += 1
    else:
        mapping[i] = 1
mapping
```

```python
month_name = []
job_posting = []
for i in mapping:
    month_name.append(i)
    job_posting.append(mapping[i])
plt.bar(month_name, job_posting)
plt.xticks(rotation = 45)
plt.show()
```

```python
location = df.location.tolist()
location
```

```python
indian_cities = []
for i in location:
    if (i[0] == 'I' and i[1] == 'N'):
        indian_cities.append(i)
indian_cities
```

```python
cities = []
for i in indian_cities:
    cities.append(i[8:])
cities
```

```python
mapping = {}
for i in cities:
    if i in mapping:
        mapping[i] += 1
    else:
        mapping[i] = 1
mapping
```

```python
city = []
percentage = []
for i in mapping:
    city.append(i)
    percentage.append(mapping[i])
plt.pie(percentage, labels = city, autopct = "%1.2f")
plt.show()
```

```python
year_total_jobs = df.filter(['Posting_date', 'BASIC QUALIFICATIONS'], axis=1)
year_total_jobs
```

```python
year_total_jobs.rename(columns = {'BASIC QUALIFICATIONS':'QUALIFICATIONS'}, inplace = True)
year_total_jobs
```

```python
year_total_jobs.dropna(subset = ["QUALIFICATIONS"], inplace=True)
year_total_jobs = year_total_jobs[year_total_jobs['QUALIFICATIONS'].str.contains('Java' or'java')]
year_total_jobs
```

```python
posting_date = year_total_jobs.Posting_date.tolist()
dates = []
for i in posting_date:
    length = len(i)
    dates.append(int(i[length - 4: length]))
mapping = {}
for i in dates:
    if i in mapping:
        mapping[i] += 1
    else:
        mapping[i] = 1
mapping = collections.OrderedDict(sorted(mapping.items()))
year = []
job_posting = []
for i in mapping:
    year.append(i)
    job_posting.append(mapping[i])
plt.scatter(year, job_posting)
plt.show()
for i in mapping:
    print(i, mapping[i])
```
